```python
import tensorflow as tf
from tensorflow.keras.layers import LayerNormalization
```

```python
class AddNorm(tf.keras.layers.Layer):
    def __init__(self, epsilon=1e-6):
        super(AddNorm,self).__init__()
        self.epsilon = epsilon
        self.layerNorm = LayerNormalization(epsilon= self.epsilon)
    def call(self, x, y):
        return self.layerNorm(x + y)
```

```python
batch_size = 16
d_model = 512
seq_len = 10
x = tf.random.normal([batch_size, seq_len, d_model])
sublayer_output = tf.random.normal([batch_size, seq_len, d_model])
add_norm_layer = AddNorm()
add_norm_layer(x, sublayer_output)
```
