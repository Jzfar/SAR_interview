动机：奇异值、特征间量纲差异，所以对数据归一化，加快模型收敛

为什么归一化能加快模型收敛？梯度更新的方向更接近最优解，各特征的梯度更接近，更新步伐均匀，避免震荡（方向和步伐都帮助避免震荡）
- 未归一化的数据各特征量纲不同，损失函数的等高线更扁平，又因为梯度更新的方向是垂直于等高线的方向，这个方向可能不指向最优解
- 各特征的梯度差异大，梯度下降时每个特征的步伐不均匀，容易引起震荡。
（为什么损失函数的等高线更扁平？因为特征量纲不同，特征分布在大值的参数小，特征分布在小值的参数大。导致损失函数的等高线在不同方向拉伸程度不同）
归一化后中，各特征映射到相同的范围，损失函数的等高线更接近圆形，各特征梯度更均匀，使得梯度更稳定，最优化的过程更快更平缓
