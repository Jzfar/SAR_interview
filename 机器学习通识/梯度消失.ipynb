梯度消失是目标函数对深层参数的导数趋于0，以至于这些参数更新缓慢或无法更新的现象。
主要原因是饱和性激活函数在某些输入区间内的梯度趋于零，在梯度反向传播的过程中，由于梯度计算的链式法则，这些较小的导数在网络中累乘，导致深层的参数的梯度非常小。
饱和性激活函数是如sigmoid、tanh等，在x趋于无穷时，y趋向于常数，导数趋于0.
非饱和的有relu、leaky relu等

```python
import numpy as np
import matplotlib.pyplot as plt
```

|函数名|饱和区间|
|------|------|
|sigmoid|绝对值大于5|
|tanh|绝对值大于5|

```python
def sigmoid(x):
    return 1 / (1 + np.exp(x))
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))
x = np.linspace(-10,10,500)
sigmoid_y = sigmoid(x)
sigmoid_dy = sigmoid_derivative(x)
plt.subplot(2, 2, 1)
plt.plot(x, sigmoid_y, label="Sigmoid", color="blue")
plt.plot(x, sigmoid_dy, label="Sigmoid Derivative", color="red", linestyle="--")
plt.title("Sigmoid and its Derivative")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
```

```python
def tanh(x):
    return np.tanh(x) 
def tanh_concrete(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
def tanh_derivative(x):
    return 1 - np.tanh(x) ** 2
tanh_y = tanh(x)
tanh_y_concrete = tanh_concrete(x)
tanh_yd = tanh_derivative(x)
plt.subplot(2, 2, 1)
plt.plot(x, tanh_y, label="tanh", color="blue")
plt.plot(x, tanh_y_concrete, label="tanh_concrete", color="red", linestyle="--")
plt.plot(x, tanh_yd, label="tanh_derivative", color="green")
plt.title("tanh, tanh_concrete and its derivative")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
```

```python
def relu(x):
    return np.where(x>0, x, 0)
def relu_derivative(x):
    return np.where(x>0, 1, 0)
relu_y = relu(x)
relu_yd = relu_derivative(x)
plt.subplot(2, 2, 1)
plt.plot(x, relu_y, label="relu", color="blue")
plt.plot(x, relu_yd, label="relu Derivative", color="red", linestyle="--")
plt.title("relu and its Derivative")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
```

```python
def leaky_relu(x, alpha=0.1):
    return np.where(x>0, x, alpha*x)
def leaky_relu_derivative(x, alpha=0.1):
    return np.where(x>0, 1, alpha)
leaky_relu_y = leaky_relu(x)
leaky_relu_yd = leaky_relu_derivative(x)
plt.subplot(2, 2, 1)
plt.plot(x, relu_y, label="leaky_relu", color="blue")
plt.plot(x, relu_yd, label="leaky_relu Derivative", color="red", linestyle="--")
plt.title("leaky_relu and its Derivative")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
```
